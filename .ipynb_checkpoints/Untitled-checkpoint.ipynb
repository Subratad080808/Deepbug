```python
import tokenize
import gensim
import pandas as pd
import ast
import astor
import pandas as pd
import os
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
import re

def flatten(input):
    new_list = []
    for i in input:
        for j in i:
            
            new_list.append(j)
            
    return new_list

def token_create(code):
    f = code.replace("\n", " ")

    data = []
    
    # iterate through each sentence in the file
    for i in sent_tokenize(f):
        temp = []

        # tokenize the sentence into words
        for j in word_tokenize(i):
            temp.append(j.lower())

        data.append(temp)
        type(data)
    return data 

prt=[]
oxc=[]
df1 = pd.DataFrame()
df2 = pd.DataFrame()

import os

path ='D:\Testing'


for root, directories, files in os.walk(path, topdown=False):
    for name in files:
        file_path =  (os.path.join(root, name))
        print(".........................................."+file_path)
        file_name = file_path
        
        with open(file_path,'rb') as f:
            tokens = tokenize.tokenize(f.readline)
            for token in tokens:
        #         print(token.string)

        #         print("line")
        #         print(type(token.line))
                yp=token.line
                idpr=token_create(yp)
        #         print(type(idpr))
        #         print("iop:::::::::::",idpr)
                yp = re.sub('\s+', ' ', yp)

        #         print(yp)
                data = []
        #         print(idpr)
        #         print(flatten(idpr))


                df2 = {"StingId":token.string,
                        "line":flatten(idpr),
                       "WITHOUT":yp

                    }


        #                             print("////////////////////////////////")
                #print(Arg_list)
                data.append(df2)

                df1 = df1.append(data, True)
        f.close()
```

```python
model = gensim.models.Word2Vec(
    window=5,
    min_count=2,
    workers=4,
)
model.build_vocab(df1['line'], progress_per=1000)
model.train(df1['line'],total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vocab=list(model.wv.key_to_index.keys())

```

```python
def get_vec(xp):
    for x in xp:
        normed_vector = model.wv.get_vector(x, norm=True)
        print("Process name:"+x)
        print(normed_vector)
```
